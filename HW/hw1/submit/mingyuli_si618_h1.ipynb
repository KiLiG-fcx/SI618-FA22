```python
import pandas as pd
import numpy as np
```

```python
# Q1 Load the dataset. (1 point)
df = pd.read_csv('book_data.csv')
```

```python
# Q2 duplicates of other records based on the book titles?
df.duplicated(subset='book_title').sum()
```

```python
df = df.drop_duplicates(subset='book_title')
```

```python
# Q2 number of unique books
len(df)
```

```python
df['author_list'] = df['book_authors'].str.split('|')
```

```python
# Q3 exactly 3 authors
df[df['author_list'].str.len()==3]['book_title'].count()
```

```python
df[df['author_list'].str.len()==3]
```

```python
page_num = []
for i in df['book_pages']:
    if i is np.NaN:
        page_num.append(np.NaN)
    else:
        page_num.append(int(i.replace(" pages",'')))
        
#page_num
```

```python
# Q4
df.insert(3,'pages',page_num)
df
```

```python
# Q5
df['pages'].median()
```

```python
# Q6
print(df['pages'].min())
print(df['pages'].max())
```

```python
# Q7
print(df[df['author_list'].str.len()==1]['pages'].mean())
print(df[df['author_list'].str.len()==2]['pages'].mean())
print(df[df['author_list'].str.len()==3]['pages'].mean())
```

```python
df[df['author_list'].str.len()>1]['pages'].mean()
```

The average page of books with 1 author is 333.35, while those with more than 1 author is 280. So having more than 1 author result in a longer book on average.
The mean number of pages of books with 1, 2, and 3 authors are 333.35, 274.78 and 277.75.

```python
# Q8
len(df[df['book_rating']>=4])
```

```python
# Q9
len(df[df['book_rating']<=4.5])
```

```python
# Q10
df['floor_rating'] = df['book_rating'].apply(np.floor)
df
```

```python
# Q11 For each of the discretized ratings (1,2,3,4,5), what is the average number of pages? (5 points)
print(df.groupby('floor_rating')['pages'].mean())
```

```python
# Q12 For each of the discretized ratings (1,2,3,4,5), what is the average number of reviews? (5 points)
print(df.groupby('floor_rating')['book_review_count'].mean())
```

```python
# Q13
df['genre_list'] = df['genres'].str.split('|')
```

```python
df2 = df.explode('genre_list').reset_index(drop=True)
df2 = df2.drop_duplicates(subset=['book_title','genre_list'])
df2
```

```python
# Q14: What is the average number of pages from different genres? What is the median? (10 points)
print(df2.groupby('genre_list')['pages'].mean())
```

```python
print(df2.groupby('genre_list')['pages'].median())
```

```python
# Q15: What is the average rating of books from different genres? What is the median? (10 points)
print(df2.groupby('genre_list')['book_rating'].mean())
```

```python
print(df2.groupby('genre_list')['book_rating'].median())
```
